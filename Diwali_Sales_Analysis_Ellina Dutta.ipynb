# **Exploratory Data Analysis**

## **Gender**

```python
#plotting a bar chart for Gender and it's count

ax = sns.countplot(x ='Gender', data = df, palette='Set2', hue = 'Gender', legend = False)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index = False) ['Amount'].sum().sort_values(by='Amount', ascending= False)
#Calculating total amount grouping by Gender
```

## **Amount**

```python
#Plotting the above in bar chart
Sales_by_gender = df.groupby(['Gender'], as_index = False) ['Amount'].sum().sort_values(by='Amount', ascending= False)

sg = sns.barplot(x='Gender',y='Amount', data=Sales_by_gender,palette='Set2', hue = 'Gender', legend = False)

for bars in sg.containers:
    sg.bar_label(bars, fmt='%.0f')
```

#### **From the above graphs we can see that not only most of the buyers are female, but also the purchasing power of the females are higher than males**

## **Age**

```python
ax = sns.countplot(data = df, x='Age Group',hue='Gender', palette='Set2')

for bars in ax.containers:
    ax.bar_label(bars)
```

#### **From the above graph we can see that across all age groups, the count of females are higher than males**

```python
# Total Sales vs Age Group
Sales_by_age = df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending =False)

sns.barplot(x='Age Group',y='Amount', data=Sales_by_age,palette='Set2', hue = 'Age Group', legend = False)
```

```python
#Total Amount vs Total Age
Sales_by_age =df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending = False)
ay = sns.barplot(x='Age Group', y='Amount', data=Sales_by_age, hue='Amount',palette='Set2', legend = False )

for bars in ay.containers:
    ay.bar_label(bars, fmt='%.0f')
```

#### **From the above graph we can see that the highest sales is achieved from the age group 26-35**

### **State**

```python
#Total no. of order from top 10 states

Sales_by_state= df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending= False).head(10)

st= sns.barplot(x='State', y='Orders',data=Sales_by_state, hue='Orders', palette='Set2', legend=False)

for bars in st.containers:
    st.bar_label(bars)

plt.xticks(rotation=45, ha='right')
plt.tight_layout()  
plt.show()
```

#### **From the above graph we can see that the Uttar Pradesh has the highest nubmer of orders**

```python
#Total sales from top 10 states

Sales_by_state= df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending= False).head(10)

st= sns.barplot(x='State', y='Amount',data=Sales_by_state, hue='Amount', palette='Set2', legend=False)

for bars in st.containers:
    st.bar_label(bars, fmt='%.0f', rotation=45, label_type='edge', padding=3)

    

plt.xticks(rotation=45, ha='right')
plt.tight_layout()  
plt.show()
```

#### **From the above graph we can see that the highest sales isfrom Uttar Pradesh**

### **Marital Status**

```python
#Count of marital status

ax = sns.countplot(data=df, x='Marital_Status', hue = 'Marital_Status', palette = 'Set2', legend=False)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#  Sales by Marital Status

sales_by_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending =False)

ax = sns.barplot(data=sales_by_state, x='Marital_Status', y= 'Amount', hue='Gender', palette='Set2')

for bars in ax.containers:
    ax.bar_label(bars, fmt='%.0f')
```

#### **From the above graphs we can see that people whose marital status is true, has highest no. of order, hence highest sale. Also, females among those have higher sales**

### **Occupation**

```python
#Total count per occupation

ax = sns.countplot(x='Occupation', data =df, hue='Occupation', palette = 'Set2')

for bars in ax.containers:
    ax.bar_label(bars, fmt='%.0f', rotation=45, label_type='edge', padding=3)

plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#Sales by Occupation
sales_by_occupation = df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

ax=sns.barplot(data=sales_by_occupation,x='Occupation', y= 'Amount',hue='Amount',palette='Set2',legend = False)

for bars in ax.containers:
    ax.bar_label(bars, fmt='%.0f', rotation =60, label_type= 'edge', padding =3)
    
plt.xticks(rotation=45,ha='right')
plt.tight_layout()
plt.show()
```

#### **From the above graph we can see that the highest sales is from the IT Sector, Healthcare and Aviation sector**

### **Product Category**

```python
#Count of each Product Category

ax=sns.countplot(data=df,x='Product_Category', palette = 'Set2', hue='Product_Category')

for bars in ax.containers:
    ax.bar_label(bars,fmt='%.0f', rotation=45,label_type='edge',padding =4)

plt.xticks(rotation=45,ha='right',)
plt.tight_layout()
plt.show()
```

#### **From the above graph we can see that the highest count is Clothing & Apparel and Food**

```python
## Sales by Product Category
sales_by_prod_cat= df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending = False)
ax=sns.barplot(data=sales_by_prod_cat,x='Product_Category',y='Amount',hue='Amount', palette='Set2',legend=False)

for bars in ax.containers:
    ax.bar_label(bars,fmt='%.0f', rotation=45, label_type='edge', padding =4)

plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
## Order by Product Category
orders_by_prod_cat= df.groupby(['Product_Category'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending = False)
ax=sns.barplot(data=orders_by_prod_cat,x='Product_Category',y='Orders',hue='Orders', palette='Set2',legend=False)

for bars in ax.containers:
    ax.bar_label(bars,fmt='%.0f', rotation=45, label_type='edge', padding =4)

plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

#### **From the above graphs we can see that the highest order and sales are achieved from is Clothing & Apparel and Food**

```python
# Top 10 most sold products

Top_10_prod= df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending = False).head(10)

ax=sns.barplot(data=Top_10_prod, x='Product_ID', y='Orders', hue='Orders', palette='Set2', legend = False)

for bars in ax.containers:
    ax.bar_label(bars,fmt='%.0f', rotation=45,label_type='edge',padding =4)

plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## **Conclusion: Married women age group 26-35 yrs from UP, Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics category**
